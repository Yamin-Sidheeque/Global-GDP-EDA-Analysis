```python
df1 = df.nlargest(10, 'gdpp')
df1
```

```python

figure, axes = plt.subplots(2, 1, sharex=True, figsize=(14, 6))
figure.suptitle('Imports and Exports of Top 7 GDPP Countries')

sns.barplot(x = df1['country'], y = df1['imports'], ax=axes[0])
sns.barplot(x = df1['country'], y = df1['exports'], ax=axes[1])
```

```python
fig, ax = plt.subplots(figsize=(14, 5))
sns.barplot(x = df1['country'], y = df1['inflation'])
```
